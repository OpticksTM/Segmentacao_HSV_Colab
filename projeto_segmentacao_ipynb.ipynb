```python
from google.colab import files
import cv2
import numpy as np
from matplotlib import pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Passo 1: Upload da imagem
print("Faa upload da imagem:")
uploaded = files.upload()

# Pegando o nome do arquivo carregado
image_name = list(uploaded.keys())[0]

# Lendo imagem em BGR e convertendo para RGB
img_bgr = cv2.imread(image_name)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

# Funo para atualizar a imagem filtrada
def update_image(h_min, h_max, s_min, s_max, v_min, v_max):
    # Cria mscara com base nos sliders HSV
    lower = np.array([h_min, s_min, v_min])
    upper = np.array([h_max, s_max, v_max])
    mask = cv2.inRange(img_hsv, lower, upper)

    # Aplica mscara na imagem RGB
    result = cv2.bitwise_and(img_rgb, img_rgb, mask=mask)

    # Mostra imagem original e filtrada lado a lado
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    plt.title("Imagem Original")
    plt.axis('off')
    plt.imshow(img_rgb)

    plt.subplot(1,2,2)
    plt.title("Imagem Filtrada HSV")
    plt.axis('off')
    plt.imshow(result)
    plt.show()

# Criando sliders
h_min_slider = widgets.IntSlider(value=0, min=0, max=179, description='H Min')
h_max_slider = widgets.IntSlider(value=179, min=0, max=179, description='H Max')
s_min_slider = widgets.IntSlider(value=0, min=0, max=255, description='S Min')
s_max_slider = widgets.IntSlider(value=255, min=0, max=255, description='S Max')
v_min_slider = widgets.IntSlider(value=0, min=0, max=255, description='V Min')
v_max_slider = widgets.IntSlider(value=255, min=0, max=255, description='V Max')

# Usando interact para ligar sliders  funo update
ui = widgets.VBox([h_min_slider, h_max_slider,
                   s_min_slider, s_max_slider,
                   v_min_slider, v_max_slider])

out = widgets.interactive_output(update_image, {
    'h_min': h_min_slider, 'h_max': h_max_slider,
    's_min': s_min_slider, 's_max': s_max_slider,
    'v_min': v_min_slider, 'v_max': v_max_slider
})

display(ui, out)
```
